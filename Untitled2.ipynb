```python
import ee

# Initialize the library (assumes you have already authenticated)
ee.Initialize(project='wids-earth-engine')

def mask_s2_clouds(image):
    """Function to mask clouds using the Sentinel-2 QA band"""
    qa = image.select('QA60')

    # Bits 10 and 11 are clouds and cirrus, respectively.
    cloud_bit_mask = 1 << 10
    cirrus_bit_mask = 1 << 11

    # Both flags should be set to zero, indicating clear conditions.
    mask = qa.bitwiseAnd(cloud_bit_mask).eq(0) \
             .And(qa.bitwiseAnd(cirrus_bit_mask).eq(0))

    return image.updateMask(mask).divide(10000)

# Load Sentinel-2 TOA reflectance data.
dataset = (ee.ImageCollection('COPERNICUS/S2')
           .filterDate('2018-01-01', '2018-01-31')
           # Pre-filter to get less cloudy granules.
           .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))
           .map(mask_s2_clouds))

# Visualization parameters
rgb_vis = {
    'min': 0.0,
    'max': 0.3,
    'bands': ['B4', 'B3', 'B2'],
}

# Note: Map.setCenter and Map.addLayer are specific to the GEE Code Editor.
# In Python (Jupyter), you would typically use the 'geemap' library to visualize.
import geemap
Map = geemap.Map()
Map.setCenter(-9.1695, 38.6917, 12)
Map.addLayer(dataset.median(), rgb_vis, 'RGB Composite')
Map
```

```python
# ==============================================================================
# PROJECT: WIDS 25 - RENEWABLE ENERGY SMART MAPPING
# STUDENT: [YOUR NAME]
# WEEK 3: IMPLEMENTATION OF FEATURE ENGINEERING & CLASSIFICATION
# ==============================================================================

# 1. SETUP & AUTHENTICATION
import ee
import geemap

# Trigger authentication flow

ee.Authenticate()
ee.Initialize(project='wids-earth-engine')



# 2. REGION OF INTEREST (ROI)
# Defining the Khandesh Corridor (Dhule/Nandurbar/Jalgaon)
# Coords: [West, South, East, North]
roi = ee.Geometry.Rectangle([74.0, 20.5, 75.5, 21.5])

# 3. DATA INGESTION (FETCHING SATELLITE DATA)
print("Fetching Sentinel-2, SRTM, and ERA5 Data...")

# A. OPTICAL DATA (Sentinel-2) - Filtered for low clouds
s2 = ee.ImageCollection('COPERNICUS/S2_SR') \
    .filterBounds(roi) \
    .filterDate('2023-01-01', '2023-05-30') \
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 5)) \
    .median() \
    .clip(roi)

# B. TOPOGRAPHIC DATA (SRTM DEM)
dem = ee.Image('USGS/SRTMGL1_003').clip(roi)
slope = ee.Terrain.slope(dem)

# C. CLIMATE DATA (ERA5-Land)
climate = ee.ImageCollection("ECMWF/ERA5_LAND/MONTHLY_AGGR") \
    .filterBounds(roi) \
    .filterDate('2023-01-01', '2023-12-31') \
    .median() \
    .clip(roi)

# 4. FEATURE ENGINEERING
print("Calculating NDVI, Wind Speed, and Slope...")

# Calculate NDVI (Vegetation Index) to filter farms
ndvi = s2.normalizedDifference(['B8', 'B4']).rename('NDVI')

# Calculate Wind Speed Magnitude from U and V components
wind_u = climate.select('u_component_of_wind_10m')
wind_v = climate.select('v_component_of_wind_10m')
wind_speed = wind_u.hypot(wind_v).rename('WindSpeed')

# Select Solar Radiation
ghi = climate.select('surface_solar_radiation_downwards').rename('GHI')

# STACKING: Combine all bands into one image for the Machine Learning model
# We use Blue(B2), Green(B3), Red(B4), NIR(B8) plus our engineered features
feature_stack = s2.select(['B2', 'B3', 'B4', 'B8']) \
    .addBands([ndvi, slope, wind_speed, ghi])

# 5. CREATING TRAINING DATA (GROUND TRUTH)
# Hardcoding coordinates for the specific sites mentioned in Week 2 report
print("Generating Training Points from Sakri & Brahmanvel...")

training_features = ee.FeatureCollection([
    # CLASS 1: SOLAR (Sakri Solar Park Coords) - Barren, Flat, PV Panels
    ee.Feature(ee.Geometry.Point([74.428, 20.985]), {'class': 1}),
    ee.Feature(ee.Geometry.Point([74.432, 20.982]), {'class': 1}),

    # CLASS 2: WIND (Brahmanvel Wind Farm Coords) - High Elevation, Ridge
    ee.Feature(ee.Geometry.Point([74.238, 21.085]), {'class': 2}),
    ee.Feature(ee.Geometry.Point([74.242, 21.088]), {'class': 2}),

    # CLASS 0: NEGATIVE (Yawal Sanctuary & Tapi River) - Forest/Water
    ee.Feature(ee.Geometry.Point([75.320, 21.350]), {'class': 0}), # Forest
    ee.Feature(ee.Geometry.Point([74.820, 21.160]), {'class': 0}), # River
    ee.Feature(ee.Geometry.Point([74.750, 20.900]), {'class': 0})  # Farm
])

# 6. TRAINING THE RANDOM FOREST MODEL
print("Training Random Forest Classifier...")

# Sample the pixels at the training points
training_data = feature_stack.sampleRegions(
    collection=training_features,
    properties=['class'],
    scale=30
)

# Initialize and Train the Classifier (100 Trees)
rf_classifier = ee.Classifier.smileRandomForest(100).train(
    features=training_data,
    classProperty='class',
    inputProperties=feature_stack.bandNames()
)

# Run the classification over the whole region
classified_map = feature_stack.classify(rf_classifier)

# 7. POST-PROCESSING (THE "SMART" FILTERS)
# Only accept Solar if Slope < 5 degrees (Engineering Constraint)
final_solar_sites = classified_map.eq(1).And(slope.lt(5))

# Only accept Wind if Wind Speed > 4 m/s (Resource Constraint)
final_wind_sites = classified_map.eq(2).And(wind_speed.gt(4))

# 8. VISUALIZATION
print("Generating Map...")
Map = geemap.Map(center=[21.1, 74.5], zoom=9)

# Add Layers
Map.addLayer(s2, {'min': 0, 'max': 3000, 'bands': ['B4', 'B3', 'B2']}, 'Satellite Image')
Map.addLayer(final_solar_sites.selfMask(), {'palette': ['orange']}, 'Predicted Solar Hotspots')
Map.addLayer(final_wind_sites.selfMask(), {'palette': ['blue']}, 'Predicted Wind Hotspots')

# Add Legend and Display
Map.add_legend(title="Suitability Legend", labels=['Solar Potential', 'Wind Potential'], colors=['orange', 'blue'])
Map
```

```python
import ee
import geemap

# Authenticate
try:
    ee.Initialize(project='wids-earth-engine')
except:
    ee.Authenticate()
    ee.Initialize(project='wids-earth-engine')

# ROI
roi = ee.Geometry.Rectangle([74.0, 20.5, 75.5, 21.5])

# NEW: Sentinel-2 cloud masking function (borrowed conceptually)
def mask_s2_sr(image):
    qa = image.select('QA60')
    cloud_bit_mask = 1 << 10
    cirrus_bit_mask = 1 << 11
    mask = qa.bitwiseAnd(cloud_bit_mask).eq(0).And(
           qa.bitwiseAnd(cirrus_bit_mask).eq(0))
    return image.updateMask(mask).divide(10000)

# Sentinel-2 data
s2 = (
    ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
    .filterBounds(roi)
    .filterDate('2023-01-01', '2023-05-30')
    .map(mask_s2_sr)      #  NEW
    .median()
    .clip(roi)
)

# DEM & slope
dem = ee.Image('USGS/SRTMGL1_003').clip(roi)
slope = ee.Terrain.slope(dem)

# ERA5
climate = (
    ee.ImageCollection("ECMWF/ERA5_LAND/MONTHLY_AGGR")
    .filterBounds(roi)
    .filterDate('2023-01-01', '2023-12-31')
    .median()
    .clip(roi)
)

# NDVI
ndvi = s2.normalizedDifference(['B8', 'B4']).rename('NDVI')

# Wind speed
wind_speed = climate.select('u_component_of_wind_10m') \
    .hypot(climate.select('v_component_of_wind_10m')) \
    .rename('WindSpeed')

# Solar radiation
ghi = climate.select('surface_solar_radiation_downwards_sum').rename('GHI')

# Feature stack
feature_stack = s2.select(['B2', 'B3', 'B4', 'B8']) \
    .addBands([ndvi, slope, wind_speed, ghi])

# OPTIONAL UPGRADE: polygon-based training (recommended)
solar_poly = ee.Geometry.Rectangle([74.42, 20.97, 74.45, 21.00])
wind_poly = ee.Geometry.Rectangle([74.22, 21.07, 74.25, 21.10])
neg_poly = ee.Geometry.Rectangle([74.75, 20.85, 74.90, 21.05])

training_fc = ee.FeatureCollection([
    ee.Feature(solar_poly, {'class': 1}),
    ee.Feature(wind_poly, {'class': 2}),
    ee.Feature(neg_poly, {'class': 0}),
])

training = feature_stack.sampleRegions(
    collection=training_fc,
    properties=['class'],
    scale=30
)

# Classifier (RF retained  best for mixed data)
classifier = ee.Classifier.smileRandomForest(100).train(
    training, 'class', feature_stack.bandNames()
)

classified = feature_stack.classify(classifier)

# Smart filtering
solar_sites = classified.eq(1).And(slope.lt(5))
wind_sites = classified.eq(2).And(wind_speed.gt(2))

# Map
Map = geemap.Map(center=[21.1, 74.5], zoom=9)
Map.addLayer(s2, {'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 0.3}, 'Sentinel-2')
Map.addLayer(solar_sites.selfMask(), {'palette': ['orange']}, 'Solar')
Map.addLayer(wind_sites.selfMask(), {'palette': ['blue']}, 'Wind')

Map.add_legend(
    title="Renewable Energy Suitability",
    labels=['Solar Potential', 'Wind Potential'],
    colors=[(255,165,0), (0,0,255)]
)

Map
```
